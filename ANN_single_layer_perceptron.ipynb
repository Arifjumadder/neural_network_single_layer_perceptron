```python
import numpy as np
import pandas as pd


# Load Dataset

data = pd.read_csv('/content/diabetes.csv')

# Input and Output
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values.reshape(-1,1)


# Normalize Data (Min-Max)

x = (x - np.min(x, axis=0)) / (np.max(x, axis=0) - np.min(x, axis=0))


# Train-Test Split (80-20)

np.random.seed(20)

indices = np.arange(len(x))
np.random.shuffle(indices)

split = int(0.8 * len(x))

train_idx = indices[:split]
test_idx = indices[split:]

x_train = x[train_idx]
y_train = y[train_idx]

x_test = x[test_idx]
y_test = y[test_idx]


# Activation Function

def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_derivative(x):
    return x*(1-x)


# Initialize Parameters

input_neurons = 8
output_neurons = 1

np.random.seed(20)

w = np.random.randn(input_neurons, output_neurons)
b = np.zeros((1, output_neurons))

learning_rate = 0.1
epochs = 5000


# Training

for epoch in range(epochs):

    net_input = np.dot(x_train, w) + b
    predicted_output = sigmoid(net_input)

    error = y_train - predicted_output

    d_output = error * sigmoid_derivative(predicted_output)

    w += np.dot(x_train.T, d_output) * learning_rate
    b += np.sum(d_output, axis=0, keepdims=True) * learning_rate

    if epoch % 1000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch}, Loss: {loss:.6f}")


# Testing

net_test = np.dot(x_test, w) + b
test_output = sigmoid(net_test)

# Convert probability into 0 or 1
predictions = (test_output >= 0.5).astype(int)


# Accuracy

correct = np.sum(predictions == y_test)
accuracy = (correct / len(y_test)) * 100

print("\nAccuracy = {:.2f}%".format(accuracy))
```
